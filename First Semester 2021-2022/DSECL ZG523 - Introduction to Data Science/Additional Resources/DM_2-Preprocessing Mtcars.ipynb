## <font color='Orchid'>Replace values with mean / mode of the feature </font> 

```python
df.columns[df.isna().any()].tolist()
```

```python
df['normalized-losses'] = pd.to_numeric(df['normalized-losses'])
df['bore'] = pd.to_numeric(df['bore'])
df['stroke'] = pd.to_numeric(df['stroke'])
df['horsepower'] = pd.to_numeric(df['horsepower'])
df['peak-rpm'] = pd.to_numeric(df['peak-rpm'])
df['price'] = pd.to_numeric(df['price'])
```

```python
mu = df['normalized-losses'].mean()
df['normalized-losses'].replace(np.NaN, mu, inplace = True)

print()
```

```python
mu = df['bore'].mean()
df['bore'].replace(np.NaN, mu, inplace = True)

print()
```

```python
mu = df['stroke'].mean()
df['stroke'].replace(np.NaN, mu, inplace = True)

print()
```

```python
mu = df['horsepower'].mean()
df['horsepower'].replace(np.NaN, mu, inplace = True)

print()
```

```python
mu = df['peak-rpm'].mean()
df['peak-rpm'].replace(np.NaN, mu, inplace = True)

print()
```

```python
mu = df['price'].mean()
df['price'].replace(np.NaN, mu, inplace = True)

print()
```

```python
df.dropna(subset=['num-of-doors'], axis=0, inplace=True)
```

```python
df.columns[df.isna().any()].tolist()
```

# <font color='CornflowerBlue'>Binning</font>

```python
bins = np.linspace( min(df['price']), max(df['price']), 4)

names = ['low', 'medium', 'high']

df['price_binned'] = pd.cut( df['price'], bins, labels = names, include_lowest = True)
```

```python
df['price_binned'].tail()
```

```python
df['price'].hist(bins=3)
plt.show()
```

```python
df['price_binned'].value_counts().plot(kind='bar')
plt.show()
```
